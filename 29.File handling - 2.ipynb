```python
import os

old_path = os.getcwd()
folder_name = input('Enter folder name: ')
os.mkdir(folder_name)
new_path = os.path.join(old_path, folder_name)
os.chdir(new_path)

start = int(input('Start: '))
end = int(input('End: '))
for num in range(start, end+1):
    file = open(f'table_of_{num}.txt', 'w')
    for i in range(1, 11):
        file.write(f"{num} * {i} = {num*i}\n")
    file.close()
    
os.chdir(old_path)
```
