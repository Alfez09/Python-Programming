```python
app_ids = []

for i in range(int(input('Enter number of query: '))):
    result = search(input('Enter your query: '), country='in', lang='en')
    for i in result:
        app_ids.append(i['appId'])
```

```python
titles = []
summary = []
installs = []
score = []
ratings = []
reviews = []
developerEmail = []
developerWebsite = []
released = []
url = []
```

```python
for i in app_ids:
    results = app(i, lang='en', country='in')
    titles.append(results['title'])
    summary.append(results['summary'])
    installs.append(results['installs'])
    score.append(results['score'])
    ratings.append(results['ratings'])
    reviews.append(results['reviews'])
    developerEmail.append(results['developerEmail'])
    developerWebsite.append(results['developerWebsite'])
    released.append(results['released'])
    url.append(results['url'])
```
