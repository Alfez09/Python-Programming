```python
class Human:

    def __init__(self,name,age,address):
        self.NAME = name
        self.AGE = age
        self.ADDRESS = address
        
    def laugh(self):
        print(f'{self.NAME} is laughing')

    def cry(self):
        print(f'{self.NAME} is crying')

    def eat(self):
        print(f'{self.NAME} is eating')

    def run(self):
        print(f'{self.NAME} is running')
```
